````python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

# خواندن تصویر
img = Image.open('/content/cameraman.jpg').convert('L')  # تبدیل به مقیاس خاکستری
img_array = np.array(img)

# اجرای FFT اول
fft_img = np.fft.fft2(img_array)
# fft_shift = np.fft.fftshift(fft_img)  # شیفت دادن برای نمایش بهتر

# اجرای FFT دوم
fft_img2 = np.fft.fft2(fft_img)
# fft_shift2 = np.fft.fftshift(fft_img2)

# نمایش نتایج
plt.subplot(1, 3, 1), plt.imshow(img_array, cmap='gray')
plt.title('Original Image')
plt.subplot(1, 3, 2), plt.imshow(np.abs(fft_img2), cmap='gray')
plt.title('FFT')
#plt.subplot(1, 3, 3), plt.imshow(np.log(1+np.abs(fft_img)), cmap='gray')
#plt.title('FFT of FFT')
plt.show()
````
